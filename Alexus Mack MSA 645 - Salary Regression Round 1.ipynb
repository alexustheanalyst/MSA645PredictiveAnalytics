```python
#Loading packages 
import pandas as pd
import statsmodels.formula.api as smf
```

```python
#Reading file and viewing data
filename = '/Users/alexu/OneDrive/Documents/MSBA/Capstone/Capstone Predictive Analytics/US Salary Survey Data Round 1.csv'
salaryData = pd.read_csv(filename)
salaryData.head()
```

```python
#Checking data types
salaryData.info()
```

```python
#Creating dummy variables for Age_group
salaryData_with_dummies = pd.get_dummies(data=salaryData, columns=['Age_group'])
```

```python
print(salaryData_with_dummies)
```

```python
#Regression expression excluding Age_group_under_18 as the intercept
reg_exp = 'Salary ~ Age_group_18_24 + Age_group_25_34 + Age_group_35_44 + Age_group_45_54 + Age_group_55_64 + Age_group_65_over'
```

```python
#Building and training the model before printing the results
olsr_model = smf.ols(formula=reg_exp, data=salaryData_with_dummies)
olsr_model_results = olsr_model.fit()
print(olsr_model_results.summary())
```
