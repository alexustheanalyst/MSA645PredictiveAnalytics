```python
#Loading packages
import pandas as pd
import statsmodels.formula.api as smf
```

```python
#Reading file and viewing data
filename = '/Users/alexu/OneDrive/Documents/MSBA/Capstone/Capstone Predictive Analytics/US Salary Survey Data Round 2.csv'
salaryData = pd.read_csv(filename)
salaryData.head()
```

```python
#Checking data types
salaryData.info()
```

```python
#Creating new dummy variables for Edu_level
salaryData_with_dummies = pd.get_dummies(data=salaryData, columns=['Edu_level'])
```

```python
print(salaryData_with_dummies)
```

```python
#Regression expression excluding High_School as the intercept
reg_exp = 'Salary ~ Edu_level_College_degree + Edu_level_Masters_degree + Edu_level_PhD + Edu_level_Professional_degree'
```

```python
#Building and training the model before printing the results
olsr_model = smf.ols(formula=reg_exp, data=salaryData_with_dummies)
olsr_model_results = olsr_model.fit()
print(olsr_model_results.summary())
```
